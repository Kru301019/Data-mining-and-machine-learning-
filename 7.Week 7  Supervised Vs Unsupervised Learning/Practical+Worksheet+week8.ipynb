<u>Selecting rows</u>

Rows in a DataFrame are selected, typically, using the iloc/loc selection methods, or using logical selectors (selecting based on the value of another column or variable).

The basic methods are:

- numeric row selection using the iloc selector.
- label-based row selection using the loc selector (this is only applicable if you have set an “index” on your dataframe.
- logical-based row selection using evaluated statements.
- Selecting rows and columns.

<i>Note that when selecting columns, if one column only is selected, the <code>.loc</code> operator returns a Series. For a single column DataFrame, use a one-element list to keep the DataFrame format</i>


```python
#numeric row selection using the iloc selector.
#select the first 7 rows

titanic.iloc[0:7, :]
```

```python
#select row number 50
titanic.loc[50, :]
```

If a squire practice is added around the row number as shown below the data will be returned as a Dataframe

If selections of a single column are made as a string, a series is returned from <code>.loc</code>, Pass a list to get a DataFrame back.


```python
titanic.loc[[50], :]
```

```python
#selection using evaluated statements.
#select rows where Embarked equal C
titanic[titanic["Embarked"] == "C"]
```

```python
#select all rows where sex equals male and display only the age column
titanic.loc[titanic["Sex"] == "male", ["Age"] ]
```

<u>Add rows</u>

In dataframe <code>.append()</code> can pass a dictionary of key-value pairs i.e.

- key = Column name.
- Value = Value at that column in a new row.

```python
titanic = titanic.append({'PassengerId' : '10000' , 'Age' : 22} , ignore_index=True)

#print new dataframe
titanic
```

The above function will not modify the existing dataframe object titanic, it will return a new dataframe containing a copy of the contents of the existing dataframe and with a new row appended at its end.

New DataFrame’s index is not the same as the original dataframe because <code>ignore_index</code> is passed as <i>True</i> in the <code>.append()</code> function. Also, for columns that were not present in the dictionary <i>NaN</i> value is added.

Add row in the dataframe using dataframe <code>.append()</code> and Series

Also a series could be passed to <code>.append()</code> to append a new row in the dataframe.

Using the following function the attribute name doesn't need to be included, just the instances should be added for each attribute as follows:


```python
# Pass a series in append() to append a row in dataframe
titanic = titanic.append(pd.Series(['900', 0, '1', 'adam K','male','30','0','0','4400','7.2','C50','C'], index=titanic.columns ), ignore_index=True)

#print new dataframe
titanic
```

Multiple rows in the dataframe could be added using dataframe <code>.append()</code> and Series.

A list of series can be passed too in dataframe <code>.append()</code> for appending multiple rows in dataframe.

The following will create a list of series with the same column names as dataframe then pass this list of series to the <code>.append()</code> function:

```python
#list of series
listOfSeries = [pd.Series(['900', 0, '3', 'adam K','male','30','0','0','4400','7.2','C7','C'], index=titanic.columns ) ,
                pd.Series(['901', 1, '1', 'justin m','male','10','0','0','4500','7.55','E50','C'], index=titanic.columns ) ,
                pd.Series(['902', 1, '1', 'sarah F','female','23','0','0','7800','54.2','C80','C'], index=titanic.columns ) ]

#print new dataframe
titanic
```

```python
# Pass a list of series to the append() to add multiple rows
titanic = titanic.append(listOfSeries , ignore_index=True)

#print new dataframe
titanic
```

<u>Renaming columns</u>

There are two ways to rename a column in DataFrame:

- Rename by mapping old names to new names using a dictionary.
- Rename by providing a function to change the column names with (Functions are applied to every column name).


<u>Rename by mapping old names to new names using a dictionary.</u>

```python
# Rename columns using a dictionary to map values
titanic = titanic.rename(columns={"Sex": "Gender"})

#print new dataframe
titanic
```

```python
#the inplace parameter will change the dataframe without assignment
titanic.rename(columns={"Pclass": "Ticket class"}, inplace=True)

#print new dataframe
titanic
```

```python
# Rename multiple columns in one go with a larger dictionary
titanic.rename(columns={"Ticket": "Ticket Number", "Cabin": "Cabin Number"}, inplace=True)

#print new dataframe
titanic
```

<u>Rename by providing a function to change the column names with. Functions are applied to every column name.</u>

```python
# Rename all columns using a function, e.g. convert all column names to lowercase
titanic.rename(columns=str.lower)

#print new dataframe
titanic
```

```python
# Rename all columns using a function, e.g. convert all column names to uppercase
titanic.rename(columns=str.upper)

#print new dataframe
titanic
```

<u>Group by</u>

Group by essentially splits the data into different groups depending on the selected variable.

The <code>groupby()</code> function returns a Groupby object but essentially describes how the rows of the original data set have been split.

Functions like </code>max()</code>, <code>min()</code>, <code>mean()</code>, <code>first()</code>, <code>last()</code>, <code>count()</code> can be applied to the Groupby object to obtain summary statistics for each group.


```python
# Get the first entry for each Embarked
titanic.groupby(['Embarked']).first()
```

```python
# Get the mean of the survived/not survived per Age
titanic.groupby(['Ticket class'])['Survived'].mean()
```

```python
#get the number of embarked
titanic.groupby(['Embarked']).count()
```
