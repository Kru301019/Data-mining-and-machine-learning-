<b>Random Forest Regression</b>

- Build the Random Forest Model
- Initialize the RF Regression model
- Fit the Model
- Prediction

```python
# import the RF Regression model from scikit learn
from sklearn.ensemble import RandomForestRegressor
```

```python
#initialize the RF regression model
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
```

```python
#fit the RF Model
regressor.fit(x_train,y_train)
```

```python
# Predicted the result
y_pred=regressor.predict(x_test)
y_pred
```

<b>Decision Tree Regression</b>

- Build the Decision Tree Model
- Initialize the Decision Tree Regression model
- Fit the Model
- Prediction

```python
# import the DT Regression model from scikit learn
from sklearn.tree import DecisionTreeRegressor
```

```python
#initialize the DT regression model
regressor=DecisionTreeRegressor(random_state=0)
```

```python
#fit the RF Model
regressor.fit(x_train,y_train)
```

```python
# Predicted the result
y_pred=regressor.predict(x_test)
y_pred
```

<b>SV Regression</b>

- Build the SVR Model
- Initialize the SV Regression model
- Fit the Model
- Prediction

```python
# import the SVR Regression model from scikit learn
from sklearn.svm import SVR
```

```python
#initialize the SVM regression model
regressor = SVR(kernel = 'linear')
```

```python
#fit the SVR Model
regressor.fit(x_train,y_train)
```

```python
# Predicted the result
y_pred=regressor.predict(x_test)
y_pred
```
