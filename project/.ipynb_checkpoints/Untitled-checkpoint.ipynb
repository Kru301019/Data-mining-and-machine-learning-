Data analysis of Hydra-Movie-Scrape

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
dt = pd.read_csv("Hydra-Movie-Scrape.csv")
```

```python
dt.head(50)
```

```python
numbers_of_records = dt.shape[0]
```

```python
numbers_of_features = dt.shape[1]
```

```python
print("Number of records:is", numbers_of_records)
print("Number of features:", numbers_of_features)
```

```python
dt.isnull().sum()
```

```python
dt['Summary'].fillna('NULL', inplace=True)
dt['Short Summary'].fillna('NULL', inplace=True)
dt['Writers'].fillna('NULL', inplace=True)
dt['YouTube Trailer'].fillna('NULL', inplace=True)
dt['Cast'].fillna('NULL', inplace=True)
```

```python
dt.isnull().sum()
```

```python
dt.describe()
```

```python
column_name = 'Year'

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=dt[column_name])
plt.title('Boxplot of ' + column_name)
plt.xlabel(column_name)
plt.show()
```

```python
column_name = 'Runtime'

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=dt[column_name])
plt.title('Boxplot of ' + column_name)
plt.xlabel(column_name)
plt.show()
```

```python
print("Maximum Runtime:", dt[column_name].max())
```

```python
# Find the index of the row with the maximum 'Runtime' value
max_index = dt['Runtime'].idxmax()

# Calculate the median of the 'Runtime' column
median_runtime = dt['Runtime'].median()

# Replace the maximum value with the medianb
dt.at[max_index, 'Runtime'] = median_runtime
```

```python
column_name = 'Runtime'

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=dt[column_name])
plt.title('Boxplot of ' + column_name)
plt.xlabel(column_name)
plt.show()
```

```python
column_name = 'Rating'

# Create a boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x=dt[column_name])
plt.title('Boxplot of ' + column_name)
plt.xlabel(column_name)
plt.show()
```

```python
dt.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Rating', data=dt, color='lightgreen')
plt.title('Box Plot of Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
x_variable = 'Year'
y_variable = 'Rating'

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(dt[x_variable], dt[y_variable], marker='o', linestyle='-', color='blue')
plt.title(f'Line Chart: {y_variable} over {x_variable}')
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.grid(True)
plt.show()
```

```python

# Specify the variables for the bar graph
x_variable = 'Rating'
y_variable = 'Runtime'  # You need to calculate the count for each genre

# Create a bar graph
plt.figure(figsize=(12, 6))
dt_grouped = dt.groupby(x_variable).size().reset_index(name=y_variable)
plt.bar(dt_grouped[x_variable], dt_grouped[y_variable], color='orange')
plt.title(f'Bar Graph: {y_variable} by {x_variable}')
plt.xlabel(x_variable)
plt.ylabel(y_variable)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Runtime', y='Rating', data=dt, color='blue', alpha=0.7)
plt.title('Scatter Plot: Rating vs. Runtime')
plt.xlabel('Runtime')
plt.ylabel('Rating')
plt.grid(True)
plt.show()
```

```python
high_rated_movies = dt[dt['Rating'] > 7.5]

# Count occurrences of each director and actor in the filtered DataFrame
director_counts = high_rated_movies['Director'].value_counts()
actor_counts = high_rated_movies['Cast'].value_counts()

# Print the director and actor with the highest counts
most_common_director = director_counts.idxmax()
count_of_most_common_director = director_counts.max()

most_common_actor = actor_counts.idxmax()
count_of_most_common_actor = actor_counts.max()

print(f"The director with the most movies rated higher than 7.5 is {most_common_director} with {count_of_most_common_director} movies.")
print(f"The actor with the most movies rated higher than 7.5 is {most_common_actor} with {count_of_most_common_actor} movies.")
```
