Min: The minimum value in each column.

25% (Percentile): The 25th percentile, or the value below which 25% of the data falls. This is also known as the first quartile.

50% (Percentile/Median): The median or 50th percentile, which is the middle value of the dataset. Half of the values are below the median, and half are above.

75% (Percentile): The 75th percentile, or the value below which 75% of the data falls. This is also known as the third quartile.

Max: The maximum value in each column.

Plot the “work class” attribute using the bar plot and value counts function.

```python
workclass_counts = adult['workclass'].value_counts()

# Plotting the bar plot
workclass_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Workclass')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```
