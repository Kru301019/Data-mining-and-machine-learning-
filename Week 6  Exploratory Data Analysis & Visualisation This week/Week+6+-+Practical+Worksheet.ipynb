<u>Data Visualization</u>

Pandas can be used to plot simple plots such as histograms, count plots, scatter plots, etc.

Pandas uses the <code>plot()</code> method to create diagrams.

```python
titanic.plot()

plt.show()
```

<u>Line plot</u>

To plot line plots with Pandas, <code>line()</code> method is used, this method uses the plot function and pass the value for x-index and y-axis as shown below:

```python
titanic.plot.line(x='Age', y='Fare', figsize=(8,6))
```

The command line above plots a line plot where the x-axis contains passengers' age and the y-axis contains the fares paid by the passengers. The <code>figsize</code> attribute could be used to change the size of the plot.

<u>Scatter Plots</u>

To plot line plots with Pandas dataframe, the <code>scatter()</code> method is used. This method similar to the line plots uses the plot function and pass the value for x-index and y-axis as shown below:

```python
titanic.plot.scatter(x='Age', y='Fare', figsize=(8,6))
```

<u>Box Plot</u>

Pandas can also be used to plot a box plot using the <code>box()</code> method.

```python
titanic.plot.box(figsize=(10,8))
```

<u>Hexagonal Plots</u>

Hexagonal plots plot the hexagons for intersecting data points on the x and y-axis. The more points intersect, the darker is the hexagon. To plot hexagonal plots with Pandas, the <code>hexbin()</code> method is used with the plot function and pass the value for x-index and y-axis as shown below:


```python
titanic.plot.hexbin(x='Age', y='Fare', gridsize=30, figsize=(8,6))

#In the output, you will see the hexagonal plot with age on x-axis and fare on y-axis.
```

<u>Kernel Density Plots</u>

the <code>kde()</code> method is used to plot kernel density plots using the plot function:

```python
titanic['Age'].plot.kde()
```

<u>Bar Plots</u>

The <code>bar()</code> to plot bar plots with Pandas using the plot function:

```python
titanic['Pclass'].value_counts().sort_index().plot.bar()
```

Also horizontal bar plots could be used with Pandas by using the following method <code>barh()</code>.

```python
titanic['Survived'].value_counts().sort_index().plot.barh()
```

When using bar plots, why did we use <i>value_counts</i> and <i>sort_index</i>? to answer the question, try the following bar() function without adding these information as shown below:

```python
titanic['Pclass'].plot.bar()
```

Based on the result above the <code>value_counts</code> and <code>sort_index</code> helped in making the data more readable this is because these functions find the count of all unique values in the given index and sort them.

<u>Histogram</u>

To draw a histogram for any column, the column name need to be specified followed by the method <code>hist()</code> method. A histogram needs only one column and it shows the frequency of each interval.

```python
titanic['Age'].hist()
```

The number of bins can be increased for the histogram using the bin attribute, as follows:

```python
titanic['Age'].hist(bins=20)
```

The style of the plots could be improved by using the <u>Seaborn library</u> and setting a value for its <code>set_style</code> attribute. For instance, set the style of the grid to dark grey using the following command:

```python
import seaborn as sns
sns.set_style('darkgrid')
```

Now again plot the histogram using the following:

```python
titanic['Age'].hist(bins=20)
#In the output, you will see dark grey grids in the background of the plot:
```

<u>Plotting multiple Attributes</u>

```python
plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
ax = sns.countplot(titanic['Survived'], hue = titanic['Sex'], palette="Greys")
plt.xticks(rotation = 30, ha='right')
```
