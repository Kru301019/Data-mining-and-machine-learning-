<u>Data Preprocessing</u>

<i>Convert Data</i>

- To convert the pandas dataframe into a list of lists.

- The Apriori library we are going to use requires the dataset to be in the form of a list of lists, where the whole dataset is a big list and each transaction in the dataset is an inner list within the outer big list. Currently, the data is in the form of a pandas dataframe.

<i>Note:</i> 

- 315 in range (0, 315) is the number of records in the dataset.
- 7 in range (0,7) is the number of columns.
- You can use <code>df.shape</code> to display these information.

```python
records = []
for i in range(0, 315):
    records.append([str(df.values[i,j]) for j in range(0, 7)])
```

<u>Applying Apriori Algorithm</u>

- min_support: The minimum support of relations (float)
- min_confidence: The minimum confidence of relations (float)
- min_lift: The minimum lift of relations (float)
- min_length: The minimum number of items in a rule
- max_length: The maximum number of items in a rule

- The optimum values for min_support,min_confidence and  min_lift arguments can be set by trying out different values and checking the association rules whether the arguments produced a valid association between items or not.


```python
# association rule apriori
association_rules = apriori(records, min_support=0.0045, min_confidence=0.7, min_lift=3, min_length=2)
association_results = list(association_rules)
```

<u>Display the Results</u>

```python
# print the total number of rules mined by the apriori class
print(len(association_results))
```

<u>Print the Rules</u>

```python
# print the first item in the association rules
print(association_results[0])
```

<u>Print Detailed Rules</u>

```python
# print the rules, the support, the confidence, and lift for each rule
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================") 
```

Download the Dataset from Moodle

<i>Description:</i> This dataset represents information about customers buying different grocery items.

<u>Import/read the dataset</u>

```python
# import dataset
df = pd.read_csv('market_basket.csv', header=None)
num_records =len(df)
print(num_records)
```

```python
# display the first few rows of the dataset
df.head()
```

<u>Dataset Description</u>

```python
df.describe()
```

<u>Display shape</u>

```python
df.shape
```

<u>Data preprocessing</u>

Convert Data

```python
records = []
for i in range(num_records):
    records.append([str(df.values[i,j]) for j in range(0, 20)])
```

<u>Applying Apriori Algorithm</u>

```python
# association rule apriori
association_rules = apriori(records, min_support=0.006, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

<u>Display the Results</u>

```python
# print the total number of rules mined by the apriori class
print(len(association_results))
```

<u>Print the rules</u>

```python
# print the first item in the association rules
print(association_results[0])
```

<u>Print Detailed Rules</u>

```python
# another way to print and display the rule, the support, the confidence, and lift for each rule
results = []
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    value0= str(items[0])
    value1= str(items[1])
    

    #second index of the inner list
    value2=str(item[1])[:7]

    #third index of the list located at 0th
    #of the third index of the inner list
    value3 =str(item[2][0][2])[:7]
    value4 =str(item[2][0][3])[:7]
    
    rows=(value0, value1, value2, value3, value4)
    results.append(rows)
    
    labels =['Title 1', 'Title 2', 'Support', 'Confidence', 'Lift']
    df= pd.DataFrame.from_records(results, columns = labels)
    print(df)
    print("=====================================")
```

Download the Dataset from Moodle

<i>Description:</i> Music data collected from a website last.fm. The dataset contains anonymized user data from the website

<u>Import/read the dataset</u>

```python
# import dataset
df = pd.read_csv('Music.csv')
```

```python
df.head()
```

<u>Print data information</u>

```python
df.describe()
```

<u>Print the shape of the data</u>

```python
df.shape
```

<u>Attributes Selection</u>

In this example the user and artists column will be used

```python
df = df[['user','artist']]
df
```

<u>Data Preprocessing</u>

Drop Duplicate Data

```python
df = df.drop_duplicates()
df.shape
```

Transform the Dataset into list

```python
records = []
for i in df['user'].unique():
    records.append(list(df[df['user'] == i]['artist'].values))
```

<u>Print type</u>

```python
print(type(records))
```

<u>Applying Apriori Algorithm</u>

```python
# association rule apriori
association_rules = apriori(records, min_support=0.01, min_confidence=0.4, min_lift=3, min_length=2)
association_results = list(association_rules)
```

<u>View the Results</u>

```python
print("There are {} Relation derived.".format(len(association_results)))
```

<u>Print the Rules</u>

```python
#print all the rules
for i in range(0, len(association_results)):
    print(association_results[i][0])
```

<u>Print Final Result with Support, Confidence, and Lift </u>

```python
# print the rules, the support, the confidence, and lift for each rule
for item in association_results:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("Rule: With " + items[0] + " you can also listen " + items[1])

    # second index of the inner list
    print("Support: " + str(item[1]))

    # third index of the list located at 0th
    # of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
